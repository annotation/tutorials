# Displaying ancient text corpora: from divergence to confluence

## Pretty and plain in the Hebrew Bible and Uruk Tablets

Text-Fabric defines a generic `pretty()` and `plain()`, but most of it was implemented in a corpus dependent way.

````python
from tf.app import use
````

We load two corpora:
* [Hebrew Bible](https://github.com/annotation/app-bhsa)
* [Uruk](https://github.com/annotation/app-uruk).

````python
A_BHSA = use("bhsa", silent="deep")
````

````python
A_URUK = use("uruk", silent="deep")
````

## BHSA: the 1001st verse

````python
A = A_BHSA
F = A.api.F
verse = F.otype.s("verse")[1000]
verse
````

````python
A.plain(verse)
````

````python
A.pretty(verse)
````

````python
A.pretty(verse, fmt="text-phono-full")
````

````python
A.pretty(verse, extraFeatures="gloss")
````

````python
A.showContext()
````

````python
A.showProvenance(jobName="demo", author="Dirk Roorda")
````

## Uruk: the 26951st line

````python
A = A_URUK
F = A.api.F
line = F.otype.s("line")[26950]
line
````

````python
A.plain(line)
````

````python
A.pretty(line)
````

````python
L = A.api.L
tablet = L.u(line, otype="tablet")[0]
A.photo(tablet)
````

````python
A.pretty(tablet)
````

````python
A.pretty(tablet, showGraphics=False)
````

````python
A.showProvenance()
````

#  Gain: towards a generic display algorithm for textual graph models

```
The     great     (his     words)     man     leapt         the window

>speaker1       speaker1<  >speaker2
        >edited-------<
                                      >damaged-<
                                                  >missing<
                  >phrase-------<                  >phrase-------------<
>phrase----------<...............>phrase---<  >phrase< >phrase---------<
>clause----------------------------------------------------------------<
>sentence--------------------------------------------------------------------------<
```

How can we box this in hierarchical HTML while preserverving the intuitive structure?

With the following requirements:

* highlighting of arbitrary nodes in several ways must be possible
* collapsing structure at arbitrary intermediate levels must be possible
* without spelling out everything to the computer:
  * autodetect reasonable defaults by inspecting the dataset as a whole
  * configure the remaining bits
* still allow hooking up special effects
